**Step 4: Open and display the test number plate photo**

```python

image=Image.open('test.jpg')
image
```

**Step 5: Extract text**

```python
extractedInformation = pytesseract.image_to_string(image,lang='eng', config='--psm 10 --oem 3 -c tessedit_char_whitelist=0123456789')
print(extractedInformation)
```
